```python
# 定义一个5行4列的矩阵A（形状为[5, 4]）
A = torch.tensor([
    [1, 2, 3, 4],   # 第0行
    [5, 6, 7, 8],   # 第1行
    [9, 10, 11, 12], # 第2行
    [13, 14, 15, 16], # 第3行
    [17, 18, 19, 20]  # 第4行
], dtype=torch.float32)  # 指定为浮点型，避免后续运算类型错误

# 定义一个4维向量x（形状为[4]，需与矩阵A的列数匹配）
x = torch.tensor([1.0, 2.0, 3.0, 4.0], dtype=torch.float32)

# 计算矩阵向量积：A * x（结果为5维向量，形状为[5]）
result = torch.mv(A, x)
result
```

```python
B=torch.ones(4,3)
B   
```

```python
torch.mm(A,B)
```
