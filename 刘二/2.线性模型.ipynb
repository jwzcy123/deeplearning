```python
# - 
x_data = np.array([1, 2, 3, 4, 5])
y_data = 2*x_data + 3 + np.random.randn(5)*0.2  # y=2x+3+


def forward(x, w, b):  #yb
    return x*w + b

def loss(x, y, w, b):  #x,y,w,b
    y_pred = forward(x, w, b)
    return (y_pred-y)**2
```

```python
# 
w_list = []
b_list = []
mse_list = []
```

```python
for w in np.arange(0,4.1,0.5):
    for b in np.arange(0,5.1,0.5):
        print(f'w={w:.1f}, b={b:.1f}')
        l_sum=0
        for x,y in zip(x_data,y_data):
            y_pred=forward(x, w, b)
            l=loss(x,y, w, b)
            l_sum+=l
            print(f'\tx:{x:.1f},y:{y:.2f},y_pred:{y_pred:.2f},l:{l:.2f}')

        mse = l_sum/len(x_data)
        w_list.append(w)
        b_list.append(b)
        mse_list.append(mse)
        print(f'MSE={mse:.3f}\n')

# 
best_idx = np.argmin(mse_list)
best_w = w_list[best_idx]
best_b = b_list[best_idx]
best_mse = mse_list[best_idx]
print('='*50)
print(f': w={best_w:.2f}, b={best_b:.2f}, MSE={best_mse:.3f}')
```

```python
# np.meshgridwb3D
w_range = np.arange(0, 4, 0.1)
b_range = np.arange(0, 5, 0.1)
W, B = np.meshgrid(w_range, b_range)

# (w,b)MSE
Z = np.zeros_like(W)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        l_sum = 0
        for x, y in zip(x_data, y_data):
            l = loss(x, y, W[i,j], B[i,j])
            l_sum += l
        Z[i,j] = l_sum / len(x_data)  # MSE

# 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(W, B, Z, cmap='viridis', alpha=0.8)

# 3D
ax.scatter(w_list, b_list, mse_list, color='red', s=20, label='Training points')

ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('MSE')
ax.set_title('MSE Loss Surface (w, b)')
ax.legend()
plt.show()
```
