```python
import torch
from torchvision import transforms
import matplotlib.pyplot as plt
from PIL import Image

# 
img = Image.open("E:\\\\\\\\hymenoptera_data\\train\\bees\\16838648_415acd9e3f.jpg")

# ToTensor  Normalize
trans_to_tensor = transforms.ToTensor()
img_tensor = trans_to_tensor(img)

trans_normalize = transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
img_normalized = trans_normalize(img_tensor)

#  numpy  (H, W, C)
img_normalized = img_normalized.numpy().transpose((1, 2, 0))  #  (H, W, C)

# 
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.title("Normalized Image")
plt.imshow(img_normalized)
plt.show()
```
