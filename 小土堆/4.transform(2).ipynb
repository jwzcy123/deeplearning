```python
import torch
from torchvision import transforms
import matplotlib.pyplot as plt
from PIL import Image

# 打开图像
img = Image.open("E:\\深度学习\\小土堆\\数据集\\hymenoptera_data\\train\\bees\\16838648_415acd9e3f.jpg")

# 定义转换操作：ToTensor 和 Normalize
trans_to_tensor = transforms.ToTensor()
img_tensor = trans_to_tensor(img)

trans_normalize = transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
img_normalized = trans_normalize(img_tensor)

# 将归一化后的图像转换为 numpy 数组并调整维度为 (H, W, C)
img_normalized = img_normalized.numpy().transpose((1, 2, 0))  # 转换维度为 (H, W, C)

# 显示原始图像和归一化后的图像
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.title("Normalized Image")
plt.imshow(img_normalized)
plt.show()
```
