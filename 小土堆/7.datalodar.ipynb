参数



dataset (Dataset) - 要从中加载数据的数据集。

batch_size (int, 可选) - 每个批次加载的样本数量（默认值为 1 ）。

shuffle (bool, 可选) - 设置为 True 可在每个轮次（epoch ）对数据重新打乱顺序（默认值为 False ）。

sampler (Sampler 或可迭代对象，可选) - 定义从数据集中抽取样本的策略。可以是实现了 __len__ 方法的任意可迭代对象。如果指定了该参数，shuffle 就不能再指定 。

batch_sampler (Sampler 或可迭代对象，可选) - 类似 sampler ，但一次返回一批索引。与 batch_size、shuffle、sampler 和 drop_last 互斥 。

num_workers (int, 可选) - 用于数据加载的子进程数量。为 0 意味着在主进程中加载数据（默认值为 0 ）。
collate_fn (可调用对象，可选) - 合并样本列表以形成张量的小批次。在从映射式数据集（map-style dataset ）进行批次加载时使用 。

pin_memory (bool, 可选) - 如果为 True ，数据加载器会在返回张量前，将其复制到 CUDA 固定内存（pinned memory ）中。如果你的数据元素是自定义类型，或者你的 collate_fn 返回的批次是自定义类型，请参见下面的示例 。

drop_last (bool, 可选) - 设置为 True 时，如果数据集大小不能被批次大小整除，就丢弃最后一个不完整的批次。如果为 False 且数据集大小不能被批次大小整除，那么最后一个批次会更小（默认值为 False ）。

timeout (数值型，可选) - 如果为正数，是从工作进程（workers ）收集一个批次的超时时间值。应始终为非负值（默认值为 0 ）。

worker_init_fn (可调用对象，可选) - 如果不为 None ，在每个工作子进程上，在设置随机种子之后、数据加载之前，会使用工作进程 ID（一个在 [0, num_workers - 1] 范围内的整数 ）作为输入调用该函数（默认值为 None ）。
