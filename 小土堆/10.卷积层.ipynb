```python
import torch
import torchvision
import matplotlib.pyplot as plt

# 定义一个转换操作，将图像数据转换为 PyTorch 张量
transform = torchvision.transforms.ToTensor()
# 加载 CIFAR-10 测试数据集，并应用上面的转换操作
dataset = torchvision.datasets.CIFAR10(root='E:\\深度学习\\小土堆\\数据集', train=False, transform=transform)
# 定义一个包含卷积层的神经网络模型
class ConvLayer(torch.nn.Module):
    def __init__(self):
        super(ConvLayer, self).__init__()
        # 定义一个卷积层：输入通道为3（RGB），输出通道为6，卷积核大小为3x3，步长为1，无填充
        self.conv = torch.nn.Conv2d(in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=0)

    def forward(self, x):
        # 在前向传播中应用卷积层
        x = self.conv(x)
        return x

# 从数据集中获取第一张图片及其标签
img, lable = dataset[0] 

# 为图片增加一个批次维度（batch dimension），以满足模型输入要求
img_input = img.unsqueeze(0)

# 创建 ConvLayer 模型的实例
tudui = ConvLayer()
# 将图片输入模型，得到输出
output = tudui(img_input)

# 准备原始图像用于显示。Matplotlib 需要 (高, 宽, 通道) 的格式
img_display = img.permute(1, 2, 0)

# 准备输出的特征图。移除批次维度并从计算图中分离出来
output_maps = output.squeeze(0).detach()

# 创建一个包含多个子图的画布
# 我们需要足够的空间来展示原始图像和6个特征图
plt.figure(figsize=(12, 5)) 

# 显示原始图像
plt.subplot(2, 4, 1) # 创建一个2x4的网格，并将此图放在第1个位置
plt.imshow(img_display)
plt.title('Original Image') # 标题：原始图像
plt.axis('off') # 隐藏坐标轴

# 循环遍历6个输出通道，并显示每一个特征图
for i in range(output_maps.shape[0]):
    plt.subplot(2, 4, i + 2) # 将后续的图依次放在第2个及之后的位置
    plt.imshow(output_maps[i], cmap='gray') # 以灰度图模式显示每个特征图
    plt.title(f'Output Channel {i}') # 标题：输出通道 i
    plt.axis('off')

plt.suptitle('Convolution: Before & After', fontsize=16) # 设置总标题：卷积前后对比
plt.tight_layout() # 自动调整子图参数，使其紧密排列
plt.show() # 显示图像
```

```python
import torch
import torchvision
import matplotlib.pyplot as plt

# --- 1. 数据准备 ---

# 定义一个转换操作，将图像数据转换为 PyTorch 张量
transform = torchvision.transforms.ToTensor()
# 加载 CIFAR-10 测试数据集，并应用上面的转换操作
dataset = torchvision.datasets.CIFAR10(root='E:\\深度学习\\小土堆\\数据集', train=False, transform=transform)
# 创建一个数据加载器，设置批次大小为64
dataloader = torch.utils.data.DataLoader(dataset, batch_size=64)

# --- 2. 模型定义 (与之前相同) ---

# 定义一个包含卷积层的神经网络模型
class ConvLayer(torch.nn.Module):
    def __init__(self):
        super(ConvLayer, self).__init__()
        # 定义一个卷积层：输入通道为3（RGB），输出通道为6，卷积核大小为3x3，步长为1，无填充
        self.conv = torch.nn.Conv2d(in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=0)

    def forward(self, x):
        # 在前向传播中应用卷积层
        x = self.conv(x)
        return x

# --- 3. 获取一个批次的数据并执行卷积 ---

# 从dataloader中获取一个迭代器
data_iter = iter(dataloader)
# 从迭代器中取出第一个批次的数据
imgs, _ = next(data_iter)  # imgs的形状是 [64, 3, 32, 32]

# 创建 ConvLayer 模型的实例
tudui = ConvLayer()
# 将整个批次的图片输入模型，得到输出
outputs = tudui(imgs)  # outputs的形状是 [64, 6, 30, 30]

# --- 4. 可视化一个批次中的前几张图片 ---

# 设置要显示的图片数量（例如批次中的前4张）
num_images_to_show = 4

# 创建一个足够大的画布来容纳所有子图
# 每张原图 + 6张特征图 = 7列。我们显示4行。
plt.figure(figsize=(15, 8)) 
plt.suptitle('Convolution: Before & After (Batch)', fontsize=16)

# 循环遍历要显示的每张图片
for i in range(num_images_to_show):
    
    # --- 显示原始图像 ---
    ax = plt.subplot(num_images_to_show, 7, i * 7 + 1)
    img_display = imgs[i].permute(1, 2, 0) # 准备第i张原始图像用于显示
    plt.imshow(img_display)
    ax.set_title(f'Original Img {i+1}')
    plt.axis('off')

    # --- 显示该图像对应的6个特征图 ---
    for j in range(6): # 6是模型的输出通道数
        ax = plt.subplot(num_images_to_show, 7, i * 7 + j + 2)
        # 从批次输出中提取第i张图片的第j个通道
        feature_map = outputs[i, j].detach() 
        plt.imshow(feature_map, cmap='gray')
        ax.set_title(f'Channel {j}')
        plt.axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96]) # 调整布局以适应总标题
plt.show() # 显示图像
```
