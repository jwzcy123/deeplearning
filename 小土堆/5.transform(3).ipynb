```python
from PIL import Image
from torchvision import transforms
```

```python
img = Image.open("E:\\深度学习\\小土堆\\数据集\\hymenoptera_data\\train\\bees\\16838648_415acd9e3f.jpg")
```

```python
trans_totensor = transforms.ToTensor()
img_tensor = trans_totensor(img)
```

```python
img.size
```

```python
trans_resize=transforms.Resize((512,512))
img_resize=trans_resize(img)#改变图片的大小
img_resize.size
```

```python
print(img_resize)
```

```python
img_resize=trans_totensor(img_resize)#把长宽改变后的图片转换为tensor对象
```

```python
trans_resize2=transforms.Resize(512)


trans_compose=transforms.Compose([trans_resize2,trans_totensor])


img_compose=trans_compose(img)
img_compose.shape

```

```python
#随机裁剪
trans_random=transforms.RandomCrop(224)#裁剪为正方形，也可以指定高宽（h,w)
trans_compose2=transforms.Compose([trans_random,trans_totensor])
img_random=trans_compose2(img)
img_random.shape
```
