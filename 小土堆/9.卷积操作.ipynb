输入图像中的数字代表这个像素中的颜色显示



卷积操作：对应位相乘再相加






stride:步长，卷积核走几步，可以是元组(H,W),控制长宽方向走几步

步长为2






```python
import torch
import torch.nn.functional as F
```

```python
input = torch.Tensor([[1,2,0,3,1],
                      [0,1,2,3,1],
                      [1,2,1,0,0],
                      [5,2,3,1,1],
                      [2,1,0,1,1]])
```

```python
kernel = torch.Tensor([[1,2,1],
                      [0,1,0],
                      [2,1,0]])
```

```python
input=torch.reshape(input,(1,1,5,5))
```

```python
kernel=torch.reshape(kernel,(1,1,3,3))
```

```python
output1=F.conv2d(input,kernel,stride=1,)
output1

```

```python
output2=F.conv2d(input,kernel,stride=2,)
output2
```

padding:填充

作用：在输入的两侧进行隐式填充（implicit padding）。

取值方式：

单个数字：对高度和宽度方向都使用相同的填充值
例：padding=1 → 上下左右都补 1 个像素

元组 (padH, padW)：分别指定高度方向和宽度方向的填充量
例：padding=(2, 3) → 高度方向上下各补 2，宽度方向左右各补 3

默认值：0（不填充）



```python
output3=F.conv2d(input,kernel,stride=1,padding=1)
output3
```
